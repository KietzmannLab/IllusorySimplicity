```python
import os
from os import path
import pickle
import numpy as np
from tqdm import tqdm

from macaquethings.plotting.default_styles import *
from macaquethings.plotting.util import legend_without_duplicate_labels

import matplotlib.pyplot as plt
import seaborn as sns

figure_style()  # set consistent plotting defaults for all figs
%matplotlib widget
```

```python
# specify general options
monkey = 'monkeyN'  # use for the whole script


arrays = list(range(1, 17))
if monkey == 'monkeyN':
    arrays.remove(6)

array_str = "_".join([str(a) for a in arrays])
sessions = "0_3_4_5" if monkey == "monkeyN" else "0_1_2_3_4_5"  # all sessions

roi_to_color = {
    'v1': 'tab:blue',
    'v4': 'tab:orange',
    'it': 'tab:green'
}

# decoding_dir = path.join('..', '..', 'results', 'decoding', f'{monkey}_stratifiedshuffle_allsess_category_vs_image')
decoding_dir = path.join('..', '..', 'results', 'decoding', f'{monkey}_stratifyimages_allsess')
```

```python
rois = [1, 2, 3]

label_info = []
roi_info = []
roi_decoding_pkls = []

for roi in rois:
    fname_cat = f"{monkey}-labels_category-sessions_{sessions}-rois_{roi}-arrays_{array_str}-baseline_0-standardize_1-lfp.pkl"
    fname_file = f"{monkey}-labels_filenames-sessions_{sessions}-rois_{roi}-arrays_{array_str}-baseline_0-standardize_1-lfp.pkl"   
    fname_animacy = f"{monkey}-labels_animate-sessions_{sessions}-rois_{roi}-arrays_{array_str}-baseline_0-standardize_1-lfp.pkl"   
    
    roi_decoding_pkls.append(fname_cat)
    roi_decoding_pkls.append(fname_file)
    roi_decoding_pkls.append(fname_animacy)
    
    label_info.append('category')
    label_info.append('stimulus identity')
    label_info.append('animacy')
    
    roi_info.append(
        {1: 'v1', 2: 'v4', 3: 'it'}[roi]
    )
    roi_info.append(
        {1: 'v1', 2: 'v4', 3: 'it'}[roi]
    )
    roi_info.append(
        {1: 'v1', 2: 'v4', 3: 'it'}[roi]
    )

print("loading roi decoding results:")
roi_decoding_results = []
for decoding_pkl in tqdm(roi_decoding_pkls):
    try:
        with open(path.join(decoding_dir, decoding_pkl), 'rb') as f:
            res = pickle.load(f)
            t = res['times']
            acc = res['accuracies']
            roi_decoding_results.append({
                'time': t,
                'accuracy': acc
            })
    except Exception as e:
        print(e)
        roi_decoding_results.append(
            {
                'time': [],
                'accuracy': []
            }
        )
```

```python
xlim = (-30, 400)
# FIG 1 --- all ROIs
fig_decoding_per_roi = plt.figure(figsize=QUARTER_PANEL_SIZE)

for roi, label, res in zip(roi_info, label_info, roi_decoding_results):
    color = roi_to_color[roi]
    plt.plot(res['time'], res['accuracy'], label=roi, color=color, linestyle='-' if label == 'category' else '--')
plt.xlabel('time (ms)')
plt.ylabel('decoding accuracy')
plt.title(f"Decoding accuracy per ROI {monkey}")
plt.xlim(xlim)
plt.legend()
plt.savefig(path.join('decoding_performance_panels', f'{monkey}_per_roi_all.svg'))

# ---

xlim = (-30, 400)
# FIG 1 --- all ROIs
fig_decoding_per_roi = plt.figure(figsize=(HALF_PANEL_SIZE[0], 1))

for roi, label, res in zip(roi_info, label_info, roi_decoding_results):
    color = roi_to_color[roi]
    #if not label == 'category':
    #    continue
    plt.plot(res['time'], res['accuracy'], label=roi, color=color, linestyle='-' if label == 'category' else '--')
plt.xlabel('time (ms)')
plt.ylabel('decoding accuracy')
plt.title(f"Decoding accuracy per ROI {monkey}")
plt.xlim(xlim)
plt.legend()
plt.savefig(path.join('decoding_performance_panels', f'{monkey}_per_roi_all_halfpage.svg'))
```

```python
xlim = (0, 300)

# FIG 1 --- all ROIs - category
fig_decoding_per_roi = plt.figure(figsize=(THIRD_WIDTH, THIRD_WIDTH / 2))

for roi, label, res in zip(roi_info, label_info, roi_decoding_results):
    color = roi_to_color[roi]
    if not label == 'category':
        continue
    plt.plot(res['time'], res['accuracy'], label=roi, color=color)
plt.xlabel('time (ms)')
plt.ylabel('decoding accuracy')
plt.title(f"Decoding accuracy per ROI {monkey} - category")
plt.xlim(xlim)
plt.legend()
plt.savefig(path.join('decoding_performance_panels', f'{monkey}_per_roi_category.svg'))

# FIG 1 --- all ROIs - filename 
fig_decoding_per_roi = plt.figure(figsize=(THIRD_PANEL_SIZE[0], 1))

for roi, label, res in zip(roi_info, label_info, roi_decoding_results):
    color = roi_to_color[roi]
    if not label == 'stimulus identity':
        continue
    plt.plot(res['time'], res['accuracy'], label=roi, color=color)
plt.xlabel('time (ms)')
plt.ylabel('decoding accuracy')
plt.title(f"Decoding accuracy per ROI {monkey} - stimulus identity")
plt.xlim(xlim)
plt.legend()
plt.savefig(path.join('decoding_performance_panels', f'{monkey}_per_roi_filenames.svg'))

# FIG 1 --- all ROIs - category
fig_decoding_per_roi = plt.figure(figsize=(THIRD_PANEL_SIZE[0], 1))

for roi, label, res in zip(roi_info, label_info, roi_decoding_results):
    color = roi_to_color[roi]
    if not label == 'animacy':
        continue
    plt.plot(res['time'], res['accuracy'], label=roi, color=color)
plt.xlabel('time (ms)')
plt.ylabel('decoding accuracy')
plt.title(f"Decoding accuracy per ROI {monkey} - animacy")
plt.xlim(xlim)
plt.legend()
plt.savefig(path.join('decoding_performance_panels', f'{monkey}_per_roi_animacy.svg'))
```
